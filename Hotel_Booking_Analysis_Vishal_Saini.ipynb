````python
# Chart - 1 visualization code
def get_count_from_column_bar(df, column_label):
  df_grpd = df[column_label].value_counts()
  df_grpd = pd.DataFrame({'index':df_grpd.index, 'count':df_grpd.values})
  return df_grpd


def plot_bar_chart_from_column(df, column_label, t1):
  df_grpd = get_count_from_column(df, column_label)
  fig, ax = plt.subplots(figsize=(10, 4))
  c= ['grey','y','r','g','red']
  ax.bar(df_grpd['index'], df_grpd['count'], width = 0.4, align = 'edge', edgecolor = 'black', linewidth = 4, color = c, linestyle = ':', alpha = 0.5)
  plt.title(t1, bbox={'facecolor':'0.8', 'pad':3})
  plt.legend()
  plt.ylabel('Count')
  plt.xticks(rotation = 15) # use to format the lable of x-axis
  plt.xlabel(column_label)
  plt.show()
````

````python
# Chart - 1 visualization code

def get_count_from_column(df, column_label):
  df_grpd = df[column_label].value_counts()
  df_grpd = pd.DataFrame({'index':df_grpd.index, 'count':df_grpd.values})
  return df_grpd

# plot a pie chart from grouped data
def plot_pie_chart_from_column(df, column_label, t1, exp):
  df_grpd = get_count_from_column(df, column_label)
  fig, ax = plt.subplots(figsize=(10,4))
  ax.pie(df_grpd.loc[:, 'count'], labels=df_grpd.loc[:, 'index'], autopct='%1.2f%%',startangle=90,shadow=True, labeldistance = 1, explode = exp)
  plt.title(t1, bbox={'facecolor':'0.8', 'pad':3})
  ax.axis('equal')
  plt.legend()
  plt.show()  
````

````python
partition_space = [0.04,0.04]
plot_pie_chart_from_column(Df, 'hotel', 'Percentage of Booking of Both the Hotels',partition_space)
````

````python
# Chart - 8 visualization 
def Month_Num(num):
  Month_dict = {'January':1, 'February':2, 'March':3, 'April':4, 'May':5, 'June':6, 'July':7, 'August':8, 'September':9, 'October':10, 'November':11, 'December':12}
  return Month_dict[num]
````

````python
df_arrival_month_resort = pd.DataFrame(Df[Df['hotel'] == 'Resort Hotel'].groupby('arrival_date_month')['hotel']. count().reset_index())
df_arrival_month_city = pd.DataFrame(Df[Df['hotel'] == 'City Hotel'].groupby('arrival_date_month')['hotel'].count().reset_index())
for i in range(len(df_arrival_month_resort['arrival_date_month'])):
  df_arrival_month_resort.iloc[i,0] = Month_Num(df_arrival_month_resort.iloc[i,0])
  df_arrival_month_city.iloc[i,0] = Month_Num(df_arrival_month_city.iloc[i,0])
````

````python
plt.figure(figsize=(13,4))
plt.plot(df_arrival_month_resort.sort_values(by=['arrival_date_month'], ignore_index = True), color='orange', label='Resort', marker = '*')
plt.plot(df_arrival_month_city.sort_values(by=['arrival_date_month'], ignore_index = True), color='black', label='City', marker = 's')
plt.title('Customer trend across the months')
x_ticks = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
plt.xticks(range(len(x_ticks)), x_ticks )
plt.legend()
plt.show()
````
